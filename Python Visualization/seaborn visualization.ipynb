```python
import pandas as pd
#to use as combo of matplot lib and seaborn
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
sns.load_dataset("tips")
```

```python
sns.load_dataset("flights")
```

```python
tips=sns.load_dataset("tips")
tips
```

```python
sns.stripplot(y="tip",data=tips,jitter=True,size=5)#size=size of dots
plt.ylabel("tips($)")
plt.show()
```

```python
sns.stripplot(y="tip",data=tips,jitter=False)
plt.ylabel("tips($)")
plt.show()
```

```python
sns.stripplot(y="total_bill",data=tips,jitter=True,size=5)#size=size of dots
plt.ylabel("bill($)")
plt.show()
```

```python
print(tips["smoker"].value_counts())
tips.groupby("smoker")["tip"].mean()
```

```python
print(tips["size"].value_counts())
tips.groupby("size")["tip"].mean()
```

```python
print(tips["sex"].value_counts())
tips.groupby("sex")["tip"].mean()
```

```python
sns.scatterplot(tips["total_bill"],tips["tip"])
```

```python
sns.scatterplot(data=tips,x="total_bill",y="tip",size=5)
```

```python
tips["day"].unique()
```

```python
sns.scatterplot(tips["total_bill"],tips["tip"],hue=tips["day"])
#hue is to plot the other value in the graph with collor variation dots
```

```python
sns.scatterplot(tips["total_bill"],tips["tip"],tips["day"])#this is working because day is already assumed as a categrical
```

```python
#without hue and dot sizes of 
sns.scatterplot(tips["total_bill"],tips["tip"],size=tips["day"])
```

```python
#with hue of dot size also
sns.scatterplot(tips["total_bill"],tips["tip"],size=tips["day"],hue=tips["day"])
```

```python
sns.scatterplot(tips["total_bill"],tips["tip"],size=tips["smoker"],hue=tips["smoker"])
```

```python
#grouping with strip plots
sns.stripplot(x="day", y="tip",data=tips)
plt.ylabel("tip $")
plt.show()
```

```python
sns.stripplot(x="sex", y="tip",data=tips)
plt.ylabel("tip $")
plt.show()
```

```python
sns.stripplot(x="smoker", y="tip",data=tips)
plt.ylabel("tip $")
plt.show()
```

```python
sns.stripplot(x="size", y="tip",data=tips)
plt.ylabel("tip $")
plt.show()
```

```python
sns.boxplot(x="tip",y="day",data=tips,orient="h")#orient h= horizont orinet v = veritical
plt.xlabel("tip $")
plt.ylabel("day")
```

```python
sns.boxplot(x="day",y="tip",data=tips,orient="v")
plt.xlabel("tip $")
plt.ylabel("day")
```

```python
sns.boxplot(x="tip",y="smoker",data=tips,orient="h")
plt.xlabel("tip $")
plt.ylabel("day")
```

```python
sns.boxplot(x="smoker",y="tip",data=tips,orient="v")
plt.xlabel("tip $")
plt.ylabel("day")
```

```python
plt.subplot(1,2,1)
sns.boxplot(x="day",y="tip",data=tips)
plt.ylabel("tips $")
plt.subplot(1,2,2)
sns.violinplot(x="day",y="total_bill",data=tips)
plt.ylabel("total bill $")
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips)
```

```python
#density plot
sns.jointplot(x="total_bill",y="tip",data=tips,kind="kde")
```
