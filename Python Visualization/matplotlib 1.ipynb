```python
x = [-3,5,7,11,13,15,17]
y = [10,2,5,6,8,7,4]

fig=plt.figure(figsize=(5,5)) 

plt.plot(x,y,"r-*")#r-* is to get points(*) and connecting lines(-)
plt.xlim(-5,20)
plt.ylim(-5,20)
plt.xlabel("Temp",size=15)
plt.ylabel("humidity",size=15)
plt.title("weather",size=10)
plt.suptitle("humidty and temperature",size=20,y=1)
```

```python
fig=plt.figure(figsize=(5,5)) 

plt.plot(x,y,"r-*")
plt.xlim(-5,20)
plt.ylim(-5,20)
plt.xlabel("Temp",size=20)
plt.ylabel("humidity",size=20)
for i,j in zip(x,y): #zip is to get the cordinates
    plt.text(i,j,"({},{})".format(i,j))#(for loop is to get the x and y cordinates of the points  in the graph)
plt.title("weather",size=15,y=1)
plt.suptitle("humidty and temperature",size=30,y=1.1)
```

```python
import numpy as np
```

```python
print(np.pi)
print(2*np.pi)
```

```python
x=np.linspace(0,2*np.pi,400)
y=np.sin(x**2)
print(x)
print(y)
print(len(x))
print(len(y))
```

```python
fig,ax=plt.subplots()
ax.plot(x,y)
ax.set_title("a single plot")
```

```python
fig,axs=plt.subplots(4)
axs[0].plot(x,y)
axs[1].plot(x,-y)
axs[2].plot(-x,-y)
axs[3].plot(-x,y)
```

```python
fig,axs=plt.subplots(2,2)
axs[0,0].plot(x,y)
axs[0,0].set_title("axis [0,0]")
axs[0,1].plot(x,-y,"tab:orange")
axs[0,1].set_title("axis [0,1]")
axs[1,1].plot(-x,-y,"tab:green")
axs[1,1].set_title("axis [1,1]")
axs[1,0].plot(-x,y,"tab:red")
axs[1,0].set_title("axis [1,0]")
```

```python
fig,axs=plt.subplots(2,2)
plt.subplots_adjust(left=0.1,bottom=0.1,right=2,top=2,wspace=0.5,hspace=0.5) #to adjust the table spaces betwwen
axs[0,0].plot(x,y)
axs[0,0].set_title("axis [0,0]")
axs[0,1].plot(x,-y,"tab:orange")
axs[0,1].set_title("axis [0,1]")
axs[1,1].plot(-x,-y,"tab:green")
axs[1,1].set_title("axis [1,1]")
axs[1,0].plot(-x,y,"tab:red")
axs[1,0].set_title("axis [1,0]")

for ax in axs.flat:
    ax.set(xlabel="linspace(0,2*np.pi,400)",ylabel="sin(x**2)") # thsi for loop is to give the wole heading for all graphs as flat
```
