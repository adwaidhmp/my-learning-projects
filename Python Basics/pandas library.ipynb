```python
#created table and give list name or heading 
print("DHAMUS SHOP")
shopl=pd.DataFrame(columns=['fruits','vegetable','snacks','dairy','hot drinks','cool drinks','icecream'])
shopl
```

```python
#defining elements in the variable
a=['apple','banana','grapes']
b=['tomato','potato','cucumber']
c=['chips','choclate','biscut']
d=['paneer','milk','yoghut']
e=['tea','coffe','blacktea']
f=['lime','shake','juice']
g=['vanila','strawberry','choclate']
```

```python
#adding elements to columns
shopl['fruits']=pd.Series(a)
shopl['vegetable']=pd.Series(b)
shopl['snacks']=pd.Series(c)
shopl['dairy']=pd.Series(d)
shopl['hot drinks']=pd.Series(e)
shopl['cool drinks']=pd.Series(f)
shopl['icecream']=pd.Series(g)
#created a index column
shopl['index']=['a','b','c']
shopl
```

```python
#giving index value as the real index
shopl['fruits']=pd.Series(a)
shopl['vegetable']=pd.Series(b)
shopl['snacks']=pd.Series(c)
shopl['dairy']=pd.Series(d)
shopl['hot drinks']=pd.Series(e)
shopl['cool drinks']=pd.Series(f)
shopl['icecream']=pd.Series(g)
shopl['index']=['a','b','c',]
shopl.set_index('index', inplace=True)
shopl
```
