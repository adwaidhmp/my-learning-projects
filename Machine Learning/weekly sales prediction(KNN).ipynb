```python
# Define independent variables (X) and dependent variable (Y)
x = df[['Store', 'Holiday_Flag', 'Temperature', 'Fuel_Price', 'CPI', 'Unemployment']]
y = df['Weekly_Sales']

print(x)
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
print(x_train)
print(x_test)
print(y_train)
print(y_test)
```

# Trying with Linear model

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
sns.kdeplot(y_pred)
sns.kdeplot(y_test)
```

```python
print(model.predict([[10,1,30,3,150,7]]))
```

```python
from sklearn.metrics import mean_absolute_percentage_error
mape=mean_absolute_percentage_error
print(mape(y_pred,y_test))
```

```python
from sklearn.metrics import r2_score
r2=r2_score(y_test,y_pred)
r2
```

# Less accuracy so using polynomial 3rd degree

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(3)

x_poly_train = poly.fit_transform(x_train)  
x_poly_test = poly.transform(x_test)

model = LinearRegression()
model.fit(x_poly_train, y_train)

y_pred = model.predict(x_poly_test)

y_pred
```

```python
sns.kdeplot(y_pred)
sns.kdeplot(y_test)
```

```python
print(model.predict(poly.transform([[10,1,30,3,150,7]])))
```

```python
from sklearn.metrics import mean_absolute_percentage_error
mape=mean_absolute_percentage_error
print(mape(y_pred,y_test))
```

```python
from sklearn.metrics import mean_squared_error, r2_score,mean_absolute_percentage_error
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)
mape=mean_absolute_percentage_error
print(mape(y_pred,y_test))
print(mse)
print(rmse)
print(r2)
```

# less r2 score so using KNN

```python
from sklearn.neighbors import KNeighborsRegressor
knn = KNeighborsRegressor(n_neighbors=3) 
knn.fit(x_train, y_train)
y_pred = knn.predict(x_test)
y_pred
```

```python
sns.kdeplot(y_pred)
sns.kdeplot(y_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score,mean_absolute_percentage_error
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)
mape=mean_absolute_percentage_error
print(mape(y_pred,y_test),"mape")
print(mse,"mse")
print(rmse,"rmse")
print(r2,"r2")
```
