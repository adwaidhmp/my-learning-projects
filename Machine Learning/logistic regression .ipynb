# Feature Selection

```python
#univarite 
import warnings
warnings.filterwarnings("ignore")

from sklearn.feature_selection import SelectKBest, f_regression

df1 = df.iloc[:,:]

result_df = pd.DataFrame(columns=['column_name', 'score'])

# Loop over all columns in the dataframe
for col_name in df1.columns:
    X = df1[[col_name]]
    y = df1['target']

    # Apply univariate feature selection to the current column
    selector = SelectKBest(score_func=f_regression, k=1)
    selector.fit(X, y)

    # Get the score of the selected feature
    score = selector.scores_[0]

    new_data = {'column_name': [col_name], 'score': [score]}
    new_df = pd.DataFrame(data=new_data)
    result_df = pd.concat([result_df, new_df], ignore_index=True)


result_df = result_df.sort_values(by='score', ascending=False)
result_df
```

```python
df.drop(columns=df[["fbs","chol","restecg","trestbps"]])
```

```python
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
print(x)
print(y)
```

```python
sns.pairplot(df,hue="target")
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.30,random_state=42)
print(x_train)
print(x_test)
print(y_train)
print(y_test)
```

```python
# normalization
```

```python
from sklearn.preprocessing import StandardScaler

# Create an instance of StandardScaler
scaler = StandardScaler()

# Fit and transform the training data
x_train = scaler.fit_transform(x_train)

# Transform the test data (using the same scaler to avoid data leakage)
x_test = scaler.fit_transform(x_test)
```

# Logistic Regression

```python
from sklearn.linear_model import LinearRegression

# Create a Linear Regression model for regression tasks
model = LinearRegression()

# Fit the model to the training data (using continuous labels in y_train)
model.fit(x_train, y_train)

# Predict on the test data
y_pred = model.predict(x_test)
```

```python
y_pred
```

# Performance Evaluation Matrix

```python
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix, ConfusionMatrixDisplay

# Assuming y_pred contains probabilities, convert it to binary labels
y_pred_class = (y_pred >= 0.5).astype(int)

# Now you can pass y_true and y_pred_class to classification_report
report = classification_report(y_test, y_pred_class)
mat = confusion_matrix(y_test, y_pred_class)
labels = ['0', '1']

print(report)
print(mat)
cm=ConfusionMatrixDisplay(mat,display_labels=labels)
cm.plot()
```
