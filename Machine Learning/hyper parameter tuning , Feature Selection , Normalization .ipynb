# Feature Selection

```python
from sklearn.feature_selection import f_classif,SelectKBest

x=df.drop(columns=["target"])
y=df["target"]

selector = SelectKBest(score_func=f_classif, k='all') 
selector.fit(x, y)

scores = selector.scores_

feature_scores = pd.DataFrame({'Feature': x.columns, 'Score': scores})
feature_scores = feature_scores.sort_values(by='Score', ascending=False)

print(feature_scores)
```

```python
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
print(x)
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
print(x_train)
print(x_test)
print(y_train)
print(y_test)
```

# Normalization

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)#to do calcutations of mean std #test data should be unseen
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
print(x_train)
print(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier(n_neighbors=5)
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix,ConfusionMatrixDisplay
report=(classification_report(y_test,y_pred))
score=accuracy_score(y_test,y_pred)
matrix=confusion_matrix(y_test,y_pred)
labels=["no","yes"]
cm=ConfusionMatrixDisplay(matrix,display_labels=labels)
print(report)
print(score)
print(matrix)
cm.plot()
```

# Hyper Parameter Tuning

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier

cls1=KNeighborsClassifier()
params={'n_neighbors':[3,5,7,9,11],'weights':['uniform','distance']}
clf=GridSearchCV(cls1,params,cv=10,scoring='accuracy')

clf.fit(x_train,y_train)

print(clf.best_params_)

clf2=KNeighborsClassifier(n_neighbors=9,weights='distance')
clf2.fit(x_train,y_train)

y_pred1=clf2.predict(x_test)
y_pred1
```

#Accuray after Tuning

```python
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix,ConfusionMatrixDisplay
labels=['1','0']
mat=confusion_matrix(y_test,y_pred1)
cm=ConfusionMatrixDisplay(mat,display_labels=labels)
report=classification_report(y_test,y_pred1)
score=accuracy_score(y_test,y_pred1)
print(report)
print(score)
print(matrix)
cm.plot()
```
